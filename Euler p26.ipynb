```python
def MultiplicativeOrder(a = None, n = None, k = 1, lastSol = 1):
    '''returns the multiplicative order of a modulo n starting a k'''
    sol = lastSol*a
    if sol%n == 1:
        return k
    else:
        k += 1
        return MultiplicativeOrder(a = a, n = n, k = k,lastSol = sol)
    
```

```python
from math import sqrt
def Primes(n):
    numbers = range(3,n+1,2)
    initial = 0
    sizeN   = n//2-1
    stopPos  =int(sqrt(n)//2)

    for indexer in numbers[0:stopPos]:
        if indexer == 0:
            continue
        for p in xrange(initial+indexer,sizeN,indexer):
            numbers[p] = 0
        initial += 1
    return [2]+ filter(None,numbers)
```

```python
primes = Primes(1000)
primes = primes[100:]
print primes
```
